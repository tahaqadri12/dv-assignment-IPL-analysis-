## (10) Does winning the toss mean winning the match?

```python
# Check if toss winner also won match
matches['toss_match_winner'] = matches['toss_winner'] == matches['winner']

# Calculate
toss_win = matches['toss_match_winner'].value_counts()
toss_win_pct = matches['toss_match_winner'].value_counts(normalize=True) * 100

# Display
print("=" * 40)
print("TOSS IMPACT ANALYSIS")
print("=" * 40)
print(f"Toss winner won match : {toss_win[True]} matches ({toss_win_pct[True]:.1f}%)")
print(f"Toss winner lost match: {toss_win[False]} matches ({toss_win_pct[False]:.1f}%)")
print("=" * 40)
```

```python
#INSIGHTS:
- Toss winners win 52.3% of matches
- Toss winners lose 47.7% of matches
- Only 5% advantage - toss is NOT a decisive factor
- Winning toss helps slightly but doesn't guarantee victory
- Team performance matters more than luck of toss
- IPL is competitive - matches not decided by toss alone
```

## (11) what was the decision taken by captain when they won the toss? 

```python
# Count toss decisions
decisions = matches['toss_decision'].value_counts()

# Display
print("Toss Decisions:")
print(f"Field: {decisions['field']} matches")
print(f"Bat: {decisions['bat']} matches")
print()

# Simple pie chart
plt.figure(figsize=(5, 5))
plt.pie(decisions.values, labels=['Field', 'Bat'], autopct='%1.1f%%')
plt.title('Toss Decisions')
plt.show()
```

```python
#INSIGHTS:
- 61.2% captains choose to field first after winning toss
- Only 38.8% choose to bat first
- Most captains prefer chasing over setting a target
- Reasons: dew factor, knowing target, pressure on batting first
- T20 format historically favors chasing teams
- Bowling first is the popular IPL strategy
```

## (15) best venus for defending and chasing.

```python
# Best defending venues
defending = matches.groupby('venue')['win_by_runs'].mean().nlargest(3)

# Best chasing venues
chasing = matches.groupby('venue')['win_by_wickets'].mean().nlargest(3)

print("Top 3 Defending Venues:")
print(defending)
print("\nTop 3 Chasing Venues:")
print(chasing)
```

```python
#INSIGHTS:
- Best defending venues: Delhi (Feroz Shah Kotla), Kolkata, Mumbai
- Best chasing venues: Mohali, Hyderabad, Jaipur
- Big grounds with large boundaries = good for defending
- Small grounds with short boundaries = good for chasing
- Wankhede and Eden Gardens good for both
- Pitch conditions also affect defending vs chasing
```

## (16) best defending and chasing teams

```python
# Best defending teams
defending_teams = matches.groupby('winner')['win_by_runs'].mean().nlargest(3)

# Best chasing teams
chasing_teams = matches.groupby('winner')['win_by_wickets'].mean().nlargest(3)

print("Best Defending Teams:")
print(defending_teams)
print("\nBest Chasing Teams:")
print(chasing_teams)
```

```python
#INSIGHTS:
- Best defending teams: Gujarat Lions, RPS, Chennai Super Kings
- Best chasing teams: Kochi Tuskers Kerala, SRH, Rajasthan Royals
- CSK is the only team good at both defending and chasing
- Defending teams have strong bowling attacks
- Chasing teams have strong batting under pressure
- CSK's balance explains their consistent success
```
